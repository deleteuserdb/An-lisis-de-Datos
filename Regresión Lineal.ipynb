```python
#importamos las librerias necesarias
import matplotlib.pyplot as plt
from sklearn import linear_model
import seaborn as sns
import numpy as np
import pandas as pd
```

```python
#almacenamos el .csv en una variable 
datos = pd.read_csv("C:/Users/juanb/OneDrive/Escritorio/Análsis de Datos/Data de Regresión Lineal.csv", sep = ",")
```

```python
#verificamos que se hayan almacenado los datos
datos
```

```python
#definimos x, y
x = datos[['metro']]  # Características (metros)
y = datos['precio']   # Etiquetas (precios)
```

```python
datos.plot.scatter(x= "metro", y= "precio")
plt.show()
```

```python
#Creamos una variable la cual almacenará la regresión lineal
regresion = linear_model.LinearRegression()

#Agregamos los datos en un Array o vector
metro = x.values.reshape((-1,1))

#creamos el modelo
modelo = regresion.fit(metro, y)

#imprimimos la intersección
print("Intersección (b)", modelo.intercept_)

#imprimimos la pendiente
print("Pendiente (m)", modelo.coef_)
```

```python
#realizamos una prediccion del precio en relacion al metro indicado
prediccion = regresion.predict([[30]])
print(f'Precio predicho para 30 metros: {prediccion[0]}')
```

```python
#creamos el gráfico de la regresión lineal
plt.scatter(x, y, label='Datos reales')
plt.plot(x, regresion.predict(metro), color='red', label='Regresión lineal')
plt.xlabel('Metros')
plt.ylabel('Precio')
plt.legend()
plt.show()
```
